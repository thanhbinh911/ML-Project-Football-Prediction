```python
import requests
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
data = requests.get(standings_url)
```

```python
from bs4 import BeautifulSoup
```

- BeautifulSoup object is used to parse HTM or XML documents and navigate, search, and modify the parse tree
- make it easier to extract and manipulate data from web pages

```python
soup = BeautifulSoup(data.text)
```

```python
soup
```

```python
#find element using css selector
standings_table = soup.select('table.stats_table')[0]# select any table elements that have the class stats_table, choose the first element which is the epl result
```

## Note: Fbref imposes rate limit on scraping to prevent abuse, blocking users who send request too quickly
-  you’ll need to implement a delay between requests to stay within their rate limit.
-  adding sleep function
-  Fbref only allows 10 requests/min, so set the time.sleep(12)
-  it takes some minutes to download everything depending your dates range

```python
standings_table
```

```python
links = standings_table.find_all('a') # search for all occurences of a specific tag in an html document
```

```python
links = [l.get("href") for l in links]
links
```

```python
links = [l for l in links if '/squads/' in l]
links
```

```python
team_urls = [f"https://fbref.com/{l}" for l in links]
team_urls
```

```python
team_url = team_urls[0]
```

```python
data = requests.get(team_url)
```

```python
import pandas as pd
from io import StringIO
```

```python
matches = pd.read_html(StringIO(data.text), match= "Scores & Fixtures") #scan all the table tagsFutureWarning: Passing literal html to 'read_html' is deprecated and will be removed in a future version. To read from a literal string, wrap it in a 'StringIO' object. and look for one has required string(or regex)
```

```python
matches = matches[0]
matches
```

```python
soup = BeautifulSoup(data.text)
```

```python
links = soup.find_all('a')
```

```python
links = [l.get("href") for l in links]
```

```python
links = [l for l in links if l and 'all_comps/shooting/' in l] # since from get method can return None for element not contain href attribute
links 
```

```python
data =  requests.get(f"https://fbref.com{links[0]}")
```

```python
shooting = pd.read_html(StringIO(data.text), match = "Shooting" )[0]
shooting.head()
```

## Clean Data

```python
shooting.columns = shooting.columns.droplevel()
```

```python
shooting.head()
```

```python
team_data = matches.merge(shooting, on = ['Date', 'Time', 'Comp', 'Round', 'Day', 'Venue', 'Result', 'GF', 'GA', 'Opponent'] )
```
