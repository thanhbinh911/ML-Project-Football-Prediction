## Scrapping data with multiple season and teams

```python
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
from io import StringIO


standings_url  = f"https://fbref.com/en/comps/9/Premier-League-Stats" # URL for Premier League standings


# Function to initialize a WebDriver instance
def initialize_driver():
    edge_options = Options()
    edge_options.add_argument("--headless")  # Run in headless mode
    return webdriver.Edge(options=edge_options)

# Function to get HTML content with Selenium and handle rate limits
def get_selenium_html(url):
    driver = initialize_driver()  # Initialize WebDriver inside the function
    try:
        driver.get(url)
        time.sleep(random.randint(6, 7))  # Add delay to avoid rate limits
        return driver.page_source
    finally:
        driver.quit()  # Ensure the WebDriver is closed after use

# Function to fetch team data
def fetch_team_data(team_url, year, standings_table):
    team_name = team_url.split('/')[-1].replace("-Stats", "").replace("-", " ")
    try:
        
        # Initialize a WebDriver for this thread
        driver = initialize_driver()
        try:
            # Get team page using Selenium
            driver.get(team_url)
            time.sleep(random.randint(6,7))
            team_html = driver.page_source
            team_soup = BeautifulSoup(team_html, 'html.parser')

            # Get matches data (Scores & Fixtures)
            matches = pd.read_html(StringIO(team_html), match="Scores & Fixtures")[0]

            # Get shooting data (if available)
            links = [l.get("href") for l in team_soup.find_all('a')]
            links = [l for l in links if l and 'all_comps/shooting/' in l]
            if links:
                try:
                    shooting_url = f"https://fbref.com{links[0]}"
                    driver.get(shooting_url)
                    time.sleep(random.randint(6,7))
                    shooting_html = driver.page_source
                    shooting = pd.read_html(StringIO(shooting_html), match="Shooting")[0]
                    shooting.columns = shooting.columns.droplevel()  # Clean up multi-level columns
                except Exception as e:
                    print(f"Error fetching shooting data for {team_name}: {e}")
                    shooting = pd.DataFrame()  # Empty DataFrame as fallback
            else:
                shooting = pd.DataFrame()  # Empty DataFrame if no links found

            # Merge matches and shooting data
            try:
                team_data = matches.merge(shooting, on = ['Comp', 'Round', 'Day', 'Venue', 'Result','Opponent'] )
            except Exception as e:
                print(f"Skipping team {team_name} due to mismatch columns: {e}")
                return None

            # Filter for Premier League data
            team_data = team_data[team_data['Comp'].str.contains('Premier League')]
            team_data['Season'] = year
            team_data['Team'] = team_name
            return team_data
        finally:
            driver.quit()  # Ensure the WebDriver is closed after use
    except Exception as e:
        print(f"Error processing team {team_name}: {e}")
        return None

# Function to scrape and process standings for each year
def scrape_standings(year):
    global standings_url
    all_matches = []
    try:
        # Use Selenium to get the standings page
        html = get_selenium_html(standings_url)
        soup = BeautifulSoup(html, 'html.parser')
        # Find the table with the class 'stats_table'
        # This is the table that contains the standings data
        standings_table = soup.select('table.stats_table')[0]
        links = [l.get("href") for l in standings_table.find_all('a')]
        links = [l for l in links if '/squads/' in l]
        team_urls = [f"https://fbref.com{l}" for l in links]

        #Get the link to the previous season
        previous_season = soup.select("a.prev")[0].get("href")
        print(previous_season)
        standings_url = f"https://fbref.com{previous_season}"


        for team_url in team_urls:
            result = fetch_team_data(team_url, year, standings_table)
            if result is not None:
                all_matches.append(result)

        # Sleep between requests to avoid rate limits
        time.sleep(random.randint(6, 7))

    except Exception as e:
        print(f"Error scraping standings for year {year}: {e}")
        time.sleep(random.randint(6, 7))  # Sleep after error before retrying

    return all_matches

# Main execution to scrape data across all years
all_data = []

years = list(range(2025, 2015, -1))  # List of seasons to scrape data from
for year in years:
    all_data.extend(scrape_standings(year))

# Convert the results into a DataFrame (if needed)
final_data = pd.concat([df for df in all_data if df is not None], ignore_index=True)
```
